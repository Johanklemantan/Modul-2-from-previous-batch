### __Nomor 4__

Disediakan 4 buah dataset (.csv) seputar persentase balita di Indonesia yang pernah mendapatkan imunisasi BCG, Campak, DPT & Polio dalam kurun tahun 1995-2017. Unduh dataset via kaggle: unduh di sini, lalu buatlah sebuah program python yang dapat menampilkan data-data tersebut dalam bentuk sebagai berikut:

Data yang missing diolah dengan metode interpolasi linear, yakni jika terdapat nilai 'n.a' akan diubah menjadi nilai antara data sebelum & sesudahnya.

Tips: gunakan method interpolate() dari pandas).

Usai program dieksekusi, akan muncul 2 buah figure sekaligus. Figure 1 berisi data persentase balita yang pernah diimunisasi, dan figure 2 berisi persentase balita yang belum pernah diimunisasi.

Setiap figure berisi 4 buah diagram batang yang menunjukkan persentase balita yang pernah diimunisasi BCG, Campak, DPT & Polio selama 1995-2017. Output yang diharapkan digambarkan sebagai berikut:

```python
bcg = pd.read_csv('Balita Terimunisasi BCG 1995-2017.csv')
campak = pd.read_csv('Balita Terimunisasi Campak 1995-2017.csv')
dpt = pd.read_csv('Balita Terimunisasi DPT 1995-2017.csv')
polio = pd.read_csv('Balita Terimunisasi Polio 1995-2017.csv')
```

```python
bcg.loc[bcg['% Balita yang pernah mendapat imunisasi BCG']=='n.a','% Balita yang pernah mendapat imunisasi BCG']=np.nan
bcg['% Balita yang pernah mendapat imunisasi BCG'] = bcg['% Balita yang pernah mendapat imunisasi BCG'].astype(float)
bcg.info()
bcg = bcg.interpolate(method='linear', limit_direction='forward')
bcg['belom']=100-bcg['% Balita yang pernah mendapat imunisasi BCG']
bcg
```

```python
polio.head(5)
polio.loc[polio['% Balita yang pernah mendapat imunisasi Polio']=='n.a','% Balita yang pernah mendapat imunisasi Polio']=np.nan
polio['% Balita yang pernah mendapat imunisasi Polio'] = polio['% Balita yang pernah mendapat imunisasi Polio'].astype(float)
polio.info()
polio=polio.interpolate(method='linear')
polio['belom']=100-polio['% Balita yang pernah mendapat imunisasi Polio']
polio
```

```python
campak.head(5)
campak.loc[campak['% Balita yang pernah mendapat imunisasi Campak']=='n.a','% Balita yang pernah mendapat imunisasi Campak']=np.nan
campak['% Balita yang pernah mendapat imunisasi Campak'] = campak['% Balita yang pernah mendapat imunisasi Campak'].astype(float)
campak.info()
campak=campak.interpolate(method='linear')
campak['belom']=100-campak['% Balita yang pernah mendapat imunisasi Campak']
campak
```

```python
dpt.head(5)
dpt.loc[dpt['% Balita yang pernah mendapat imunisasi DPT']=='n.a','% Balita yang pernah mendapat imunisasi DPT']=np.nan
dpt['% Balita yang pernah mendapat imunisasi DPT'] = dpt['% Balita yang pernah mendapat imunisasi DPT'].astype(float)
dpt.info()
dpt=dpt.interpolate(method='linear')
dpt['belom']=100-dpt['% Balita yang pernah mendapat imunisasi DPT']
dpt
```

```python
plt.figure(figsize=(10,8))
plt.subplot(221)
plt.title('BCG',fontsize = 7)
plt.bar(bcg['Tahun'],bcg['% Balita yang pernah mendapat imunisasi BCG'],color='r')
plt.xticks(np.arange(1995,2018),rotation='vertical',fontsize = 7)

plt.subplot(222)
plt.title('Campak',fontsize = 7)
plt.bar(campak['Tahun'],campak['% Balita yang pernah mendapat imunisasi Campak'],color='g')
plt.xticks(np.arange(1995,2018),rotation='vertical',fontsize = 7)
print('')
plt.subplot(223)
plt.title('DPT',fontsize = 7)
plt.bar(dpt['Tahun'],dpt['% Balita yang pernah mendapat imunisasi DPT'],color='y')
plt.xticks(np.arange(1995,2018),rotation='vertical',fontsize = 7)

plt.subplot(224)
plt.title('Polio',fontsize = 7)
plt.bar(polio['Tahun'],polio['% Balita yang pernah mendapat imunisasi Polio'],color='b')
plt.xticks(np.arange(1995,2018),rotation='vertical',fontsize = 7)
plt.show()
```
