### __1.Populasi ASEAN Bar Chart__ 

Buatlah sebuah file python yang dapat mengakses daftar negara ASEAN dari database world dan menampilkan populasi negara-negara ASEAN dalam bentuk diagram batang. Output yang diharapkan berupa diagram batang dengan nilai/jumlah populasi tertera di atas diagram. Style, desain & warna tidak harus sama, utamakan validitas data:

```python
import mysql.connector
import sqlalchemy as db
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
con = mysql.connector.connect(
    host = 'localhost',
    user = 'root',
    passwd = 'Funicula8'
)
```

```python
c = con.cursor(buffered=True)
query = 'show databases'
c.execute(query)
```

```python
c.fetchall()
```

```python
query = 'USE world'
c.execute(query)
```

```python
query = "SELECT co.Name AS Negara_ASEAN, co.Population AS Populasi_Negara \
FROM country co \
JOIN city ci \
ON ci.CountryCode = co.Code \
WHERE ci.Name IN ('Bandar Seri Begawan','Phnom Penh','Dili','Jakarta','Vientiane','Kuala Lumpur','Rangoon (Yangon)','Manila','Singapore','Bangkok','Hanoi') \
ORDER BY Negara_ASEAN ASC;"
c.execute(query)
res = c.fetchall()
res
```

```python
haha = pd.DataFrame(res)
haha.columns=(['Negara','Populasi (x100juta) jiwa'])
haha
```

```python
plt.figure(figsize=(10,6))
plt.title('Populasi Negara ASEAN')
plt.xticks(rotation=45)
a=sns.barplot(data = haha,x='Negara',y='Populasi (x100juta) jiwa',palette = 'prism')

for i in range(len(haha.index)):
    plt.annotate(((haha['Populasi (x100juta) jiwa'][i])), xy=(haha.index[i],haha['Populasi (x100juta) jiwa'][i]), ha='center', va='bottom')
plt.grid()
plt.show()
```

### __2. Pie Chart__
Buatlah sebuah file python yang dapat mengakses daftar negara ASEAN dari database world dan menampilkan persentase populasi negara-negara ASEAN dalam bentuk diagram lingkaran. Output yang diharapkan berupa diagram lingkaran dengan nilai persentase populasi tertera dalam diagram. Style, desain & warna tidak harus sama, utamakan validitas data:

```python
plt.figure(figsize=(15,9))
plt.title('Presentase Penduduk ASEAN',fontsize = 25)
plt.pie(haha['Populasi (x100juta) jiwa'],labels=haha['Negara'],
       autopct='%.1f%%',textprops={'color':"k"})
plt.show()
```

### __3.Bar Lagi Pendapatan Bruto Nasional__
Buatlah sebuah file python yang dapat mengakses daftar negara ASEAN dari database world dan menampilkan Pendapatan Bruto Nasional/GNP (Gross National Product) negara-negara ASEAN dalam bentuk diagram batang. Output yang diharapkan berupa diagram batang dengan nilai GNP tertera di atas diagram. Style, desain & warna tidak harus sama, utamakan validitas data:

```python
query = "SELECT co.Name AS Negara_ASEAN, co.GNP \
FROM country co \
JOIN city ci \
ON ci.CountryCode = co.Code \
WHERE ci.Name IN ('Bandar Seri Begawan','Phnom Penh','Dili','Jakarta','Vientiane','Kuala Lumpur','Rangoon (Yangon)','Manila','Singapore','Bangkok','Hanoi') \
ORDER BY Negara_ASEAN ASC;"
c.execute(query)
result = c.fetchall()
result
hehe = pd.DataFrame(result)
hehe.columns=('Negara','Gross National Product (US$)')
hehe
```

```python
plt.figure(figsize=(10,6))
plt.title('Pendapatan Bruto Nasional ASEAN')
plt.xticks(rotation=45)
a=sns.barplot(data = hehe,x='Negara',y='Gross National Product (US$)',palette = 'prism')

for i in range(len(hehe.index)):
    plt.annotate(((hehe['Gross National Product (US$)'][i])), xy=(hehe.index[i],hehe['Gross National Product (US$)'][i]), ha='center', va='bottom')
plt.grid()
plt.show()
```

### __4. Pie Chart Lagi Presentase Luas Daratan ASEAN__

```python
def bulan(x):
    return x.split('-')[1]
```
