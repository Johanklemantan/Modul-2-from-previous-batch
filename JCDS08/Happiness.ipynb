### __2.1__
Untuk setiap numerik feature/column (kecuali column Overall Rank), hitunglah:

Mean, Median dan Modus
Range, Q1, Q3 dan IQR
Standard Deviasi & Variance
Z-score tiap data point

```python
# no1 = pd.DataFrame()
# hehe = []
# for x,y in enumerate (a):
# #     print(x)
#     hehe.append(a[x])
# print(hehe)

cb_mean=[]
cb_median=[]
cb_range=[]
cb_Q1 = []
cb_Q3=[]
cb_IQR=[]
cb_std=[]
cb_var=[]
cb_zscore=[]
# cb_mode=[]
for i in data.columns[2:]:

    cb_mean.append(data[i].mean())
    cb_median.append(data[i].median())
    cb_var.append(data[i].var())
    cb_range.append(np.ptp(data[i]))
    cb_std.append(data[i].std())
    cb_IQR.append((np.quantile(data[i],.75))-(np.quantile(data[i],.25)))
    cb_Q1.append(np.quantile(data[i],.25))
    cb_Q3.append(np.quantile(data[i],.75))
    cb_zscore.append(stats.zscore(data[i]))
# # for x,y in enumerate (a):

nomer1 = pd.DataFrame([cb_mean,cb_median, ide,cb_range,cb_Q1,cb_Q3,cb_IQR,cb_std,cb_var,cb_zscore], columns=data.columns[2:])
nomer1.rename({0:'mean',1:'median',2:'mode',3:'range',4:'Q1',5:'Q3',6:'IQR',7:'std dev',8:'variasi',9:'Z score tiap data'},inplace=True)
nomer1
```

### __IQR Method__

```python
cb_Q1 = []
cb_Q3=[]
cb_IQR=[]
minimum=[]
maximum=[]
for i in data.columns[2:]:
    cb_IQR.append((np.quantile(data[i],.75))-(np.quantile(data[i],.25)))
    cb_Q1.append(np.quantile(data[i],.25))
    cb_Q3.append(np.quantile(data[i],.75))
    minimum.append((np.quantile(data[i],.25))-(1.5*((np.quantile(data[i],.75))-(np.quantile(data[i],.25)))))
    maximum.append((np.quantile(data[i],.75))+(1.5*((np.quantile(data[i],.75))-(np.quantile(data[i],.25)))))
nomer2 = pd.DataFrame([cb_Q1,cb_Q3,cb_IQR,minimum,maximum], columns=data.columns[2:])
nomer2.rename({0:'Q1',1:'Q3',2:'IQR',3:'Minimum',4:'Maximum'},inplace=True)
nomer2
```

```python
no2 = pd.DataFrame()
for i in data.columns[2:]:
    hasil = []
    hasil.append((np.quantile(data[i],.75))-(np.quantile(data[i],.25)))
    hasil.append(np.quantile(data[i],.25))
    hasil.append(np.quantile(data[i],.75))
    hasil.append((np.quantile(data[i],.25))-(1.5*((np.quantile(data[i],.75))-(np.quantile(data[i],.25)))))
    hasil.append((np.quantile(data[i],.75))+(1.5*((np.quantile(data[i],.75))-(np.quantile(data[i],.25)))))
    no2[i] = hasil
print(hasil)   
no2.rename({0:'Q1',1:'Q3',2:'IQR',3:'Minimum',4:'Maximum'},inplace=True)
no2
```

### __Covariance__

```python
happy=[]
for i in data['Score']:
    happy.append(i-no2b['Score'][0])
    
gdp=[]
for i in data['GDP per capita']:
    gdp.append(i-no2b['GDP per capita'][0])

social=[]
for i in data['Social support']:
    social.append(i-no2b['Social support'][0])

health=[]
for i in data['Healthy life expectancy']:
    health.append(i-no2b['Healthy life expectancy'][0])

free=[]
for i in data['Freedom to make life choices']:
    free.append(i-no2b['Freedom to make life choices'][0])
    
gen=[]
for i in data['Generosity']:
    gen.append(i-no2b['Generosity'][0])
    
corr=[]
for i in data['Perceptions of corruption']:
    corr.append(i-no2b['Perceptions of corruption'][0])
```

```python
data['GDP per capita']
```

```python
no3a = pd.DataFrame()
no3a['Score']=happy
no3a['GDP']=gdp
no3a['GDP_Score']=no3a['Score']*no3a['GDP']
no3a['social']=social
no3a['social_Score']=no3a['Score']*no3a['social']
no3a['health']=health
no3a['health_Score']=no3a['Score']*no3a['health']
no3a['free']=free
no3a['free_Score']=no3a['Score']*no3a['free']
no3a['gen']=gen
no3a['gen_Score']=no3a['Score']*no3a['gen']
no3a['corr']=corr
no3a['corr_Score']=no3a['Score']*no3a['corr']
no3a
```

```python
no3aa = no3a[['GDP_Score','social_Score','health_Score','free_Score','gen_Score','corr_Score']]
no3aa
```

```python
pd.set_option('display.max_rows', 10)
no3aa['corr_Score'].sum()
```

```python
covariance=[]
gdp1=0
for i in no3aa['GDP_Score']:
    gdp1+=i
covariance.append(gdp1)

soc1=0
for i in no3aa['social_Score']:
    soc1+=i
covariance.append(soc1)    

hea1=0
for i in no3aa['health_Score']:
    hea1+=i
covariance.append(hea1) 

free1=0
for i in no3aa['free_Score']:
    free1+=i
covariance.append(free1) 

gen1=0
for i in no3aa['gen_Score']:
    gen1+=i
covariance.append(gen1) 

corr1=0
for i in no3aa['corr_Score']:
    corr1+=i
covariance.append(corr1) 

    
covariance
```

### __Pearson__

```python
no3b = pd.DataFrame()
no3b['Score']=happy
no3b['GDP']=gdp
no3b['social']=social
no3b['health']=health
no3b['free']=free
no3b['gen']=gen
no3b['corr']=corr
no3b['Score2']=no3b['Score']**2
no3b['GDP2']=no3b['GDP']**2
no3b['social2']=no3b['social']**2
no3b['health2']=no3b['health']**2
no3b['free2']=no3b['free']**2
no3b['gen2']=no3b['gen']**2
no3b['corr2']=no3b['corr']**2
no3b
```
